```python
def fib(n):
    results = []
    a,b = 0,1
    while a<n:
        results.append(a)
        a,b = b,b+a
    return results

out = fib(20)
print(out)
```

```python
def ask_ok(prompt,retries=4, reminder="Please try again!!"):
    while True:
        ok = input(prompt)
        if ok in ('y', 'ye', 'yes'):
            return True
        if ok in ('n', 'no', 'nop', 'nope'):
            return False
        retries = retries - 1
        if retries < 0:
            raise ValueError('Invalid user response!')
        print(reminder)

```

```python
#Lets create a class to iterate through char in string literals and then print the reverse of those characters
class Reverse:
    #We initialise the attributes of this class Reverse:
    def __init__(self, data):
        self.data = data
        self.index = len(data)
    #We create the iter method
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.index == 0: #If the self.index is equivalent to zero 
            raise StopIteration
        self.index = self.index - 1
        return self.data[self.index]
```

```python
#Create an instance of class Reverse and pass in the arguments
rev = Reverse("Spam")
```

```python
iter(rev)
```

```python
#Create a for loop to print out every char in the string literal in reverse order
for char in rev:
    print(char)
```

```python
#we are using generators to create iterators 
#Generators are a simple but powerful tools for creating iterators, they are written like regular functions but use the yield keyword whenever they want to 
#return data.
```

```python
#A typical example of a generator would be:

def Reverse(data):
    for index in range(len(data)-1,-1,-1):
        yield data[index]
```

```python
rev = Reverse("Data")
```

```python
for char in rev:
    print(char)
```
